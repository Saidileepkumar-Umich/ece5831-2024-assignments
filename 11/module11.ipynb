```python
from imdb import Imdb
from reuters import Reuters
from boston_housing import BostonHousing
```

```python

# Binary Classification
imdb = Imdb()
imdb.prepare_data()
imdb.build_model()
imdb.train()
imdb.plot_loss()
imdb.plot_accuracy()
print("IMDB Evaluation:", imdb.evaluate())
```

```python

# Multiclass Classification
reuters = Reuters()
reuters.prepare_data()
reuters.build_model()
reuters.train()
reuters.plot_loss()
reuters.plot_accuracy()
print("Reuters Evaluation:", reuters.evaluate())
```

```python

# Regression
boston = BostonHousing()
boston.prepare_data()
boston.build_model()
boston.train()
boston.plot_loss()
print("Boston Housing Evaluation:", boston.evaluate())
```
