# Module 5-3 Notebook
This notebook demonstrates using `mnist.py` and `module5-3.py` to test handwritten images.

Each image is processed by `module5-3.py`, and results are displayed below.

```python
import os
import matplotlib.pyplot as plt

image_dir = "mnist_samples"  

images = [f for f in os.listdir(image_dir) if f.endswith('.png')]

for image in images:
    actual_digit = int(image.split("_")[0])  # Extract the digit from filename

    image_path = os.path.join(image_dir, image)

    img = plt.imread(image_path)
    plt.imshow(img, cmap='gray')
    plt.title(f"Testing Image: {image}")
    plt.axis('off')
    plt.show()

    print(f"Running module5-3.py on {image} with actual digit {actual_digit}")
    
    !python module5-3.py "{image_path}" "{actual_digit}"
    print("\n")
```
